```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('SQL Stock Data Clean.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
# Replacing our missing values and fillinf values with 0
df[['Natural_Gas_Vol.', 'Crude_oil_Vol.', 'Copper_Vol.', 'Platinum_Vol.', 'Silver_Vol.', 'Gold_Vol.']] = df[['Natural_Gas_Vol.', 'Crude_oil_Vol.', 'Copper_Vol.', 'Platinum_Vol.', 'Silver_Vol.', 'Gold_Vol.']].fillna(0)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
# Changing Date From 'Object' to 'Datetime' Datatype
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
```

```python
df.info()
```

```python
# Changing columns that are object data type to float data type and if our object data has a comma it will be removed to convert our data type to float
for col in df.columns:
    if df[col].dtype == object:
        df[col] = df[col].str.replace(',','').astype(float)
```
